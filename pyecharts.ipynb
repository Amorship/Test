```python
#有空再研究
title = '正負數長條圖'
index = pd.date_range('6/27/2019', periods=6, freq='M')
df1 = pd.DataFrame(np.random.randn(6), index=index) 
df2 = pd.DataFrame(np.random.randn(6), index=index) 

dtvalue1 = [i[0] for i in df1.values]
dtvalue2 = [i[0] for i in df1.values]
_index = [i for i in df1.index.format()]

bar = Bar(title, 'Profit and loss situation')
bar.add('利潤', _index, dtvalue1)
bar.add('損失', _index, dtvalue2)
bar
```

# 漏斗圖

```python
import pyecharts
pyecharts.Funnel?
```

```python
from pyecharts import Funnel
```

```python
attr = ['GHKR', 'JG', 'WZ', 'WZHK']
value = [10, 20, 30, 40]
funnel = Funnel("漏斗圖", width=600, height=400, title_pos='center')

funnel.add("遊戲類別輪廓數據", attr, value, is_label_show=True,label_formatter='{b} {c}',
           
           label_pos="inside", legend_orient='vertical', legend_pos='left')

#legend_orient、legend_pos 分别表示圖的方向和位置
#label_formatter 標籤顯示方法 {a}種類名稱，{b}細項名稱，{c}數值, {d}數值百分比
#lable_text_color='#fff' 好像可以換顏色 待確認
#label_pos='outside' 則細項名稱會在外面
```

```python
funnel = Funnel("轉換率", "比例表示該種類下的遊戲用戶占總用戶數（100）的比例",width=600, height=400, title_pos='center')
 
funnel.add("遊戲類別輪廓數據", attr, [int(100*i/100) for i in value], is_label_show=True,
           
           label_formatter='{b} {c}%', label_pos="outside", legend_orient='vertical', legend_pos='left')

#label_formatter = {d} 則顯示方法不同 (有打在下一格)
```

```python
funnel = Funnel("轉換率", "比例表示該種類下的遊戲用戶占總用戶數（100）的比例",width=600, height=400, title_pos='center')
 
funnel.add("遊戲類別輪廓數據", attr, value, is_label_show=True,
           
           label_formatter='{b} {d}%', label_pos="outside", legend_orient='vertical', legend_pos='left')

#感覺上是較好的, 切換三種類別的時候%數會跟著跑
```
